```python
import pandas as pd
import numpy as np

# Muat Dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)

# Eksplorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum())

# Filtering - Penumpang Wanita Kelas 1
wk1 = df[(df['Sex']=='female') &
 (df['Pclass']==1)]
print(wk1['Survived'].mean(), wk1['Age'].mean())

# Groupby - Tingkat Keselamatan per Kelas
hasil = df.groupby('Pclass').agg(
    total=('PassengerId','count'),
    selamat=('Survived','mean')
).round(3)
print(hasil)

# Presentase penumpang yang selamat
selamat = np.sum(df['Survived']) / len(df) * 100
print(f"\nSurvived = {selamat:2f}%")

# Jumlah penumpang kelas 1
jumlah_wk1 = len(wk1)
print(f"Penumpang Wanita Kelas 1 = {jumlah_wk1}")

Survival_by_sex = df.groupby("Sex")["Survived"].mean() * 100

print("\nTingkat Keselamatan Berdasarkan Jenis Kelamin:")

print(f"Female : {Survival_by_sex['female']:.2f}%")
print(f"Male   : {Survival_by_sex['male']:.2f}%")
```
